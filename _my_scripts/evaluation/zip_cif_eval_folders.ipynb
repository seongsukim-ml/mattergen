```python
import pandas as pd
from pymatgen.io.cif import CifParser
from pyxtal import pyxtal
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
import os
from pathlib import Path
import zipfile
from tqdm import tqdm
```

```python
def get_symmetry_match_rate(cif_path,spg):
    cif_list = [f for f in os.listdir(cif_path) if f.endswith('.cif')]
    cif_list.sort(key=lambda x: int(x.split('_')[1].split('.')[0]))
    cif_list = [os.path.join(cif_path, f) for f in cif_list]
    
    structures = []
    try:
        for cif_file in cif_list:
            parser = CifParser(cif_file)
            structure = parser.get_structures(primitive=False)[0]
            structures.append(structure)
    except Exception as e:
        print(e)
        print("Failed to parse CIF files")
        return

    def save_res(tol):
        res = {}
        os.makedirs(cif_path.with_name("generated_crystals_cif_spg"), exist_ok=True)
        for i,structure in enumerate(structures):
            try:
                sgn = SpacegroupAnalyzer(structure,symprec=tol).get_space_group_number()
            except:
                sgn = 0
            if sgn == spg and tol == 0.1:
                structure.to(
                    filename=cif_path.with_name("generated_crystals_cif_spg") / f"spg{spg}_{i}.cif"
                )
            # print(i,sgn)
            if sgn not in res:
                res[sgn] = 0
            res[sgn] += 1
        # for k in sorted(res.keys()):
        #     print(f"{k:3} {res[k]}  {res[k]/len(structures):.2f}")

        # if spg in res:
        #     print(f"{spg} {res[spg]}  {res[spg]/len(structures):.2f}")
        # else:
        #     print(f"spg{spg} is not found")

        if spg not in res:
            res[spg] = 0
        
                # write the number of spg
        # write the number of spg
        with open(
            cif_path.parent / f"symprec_{tol:.2f}_spg{spg}_{res[spg]}_{len(structures)}.txt", "w"
        ) as f:
            for k in sorted(res.keys()):
                f.write(f"{k:3} {res[k]}  {res[k]/len(structures):.2f}\n")
            f.write("\n")
            if spg in res:
                f.write(f"{spg} {res[spg]}  {res[spg]/len(structures):.2f}\n")
            else:
                f.write(f"spg{spg} is not found\n")
    try:
        save_res(0.01)
    except Exception as e:
        print(e)
        print("Failed to save with symprec=0.01")
        with open(cif_path.parent / f"symprec_{0.01}_spg{spg}_failed.txt", "w") as f:
            f.write(str(e))
    try:
        save_res(0.05)
    except Exception as e:
        print(e)
        print("Failed to save with symprec=0.05")
        with open(cif_path.parent / f"symprec_{0.05}_spg{spg}_failed.txt", "w") as f:
            f.write(str(e))
    try:
        save_res(0.1)
    except Exception as e:
        print(e)
        print("Failed to save with symprec=0.10")
        with open(cif_path.parent / f"symprec_{0.10}_spg{spg}_failed.txt", "w") as f:
            f.write(str(e))
```

```python
dir_path = "/home/holywater2/crystal_gen/mattergen/results/0210_clip6"
dir_list = os.listdir(dir_path)
dir_list.sort()
```

```python
for i in tqdm(range(len(dir_list))):
    current_path = Path(dir_path) / dir_list[i] / "result_0"
    # print(current_path)
    if (current_path / "generated_crystals_cif").exists() and not (current_path / "generated_crystals_cif_spg").exists():
        spg = int(dir_list[i].split("_")[0].replace("spg", ""))
        get_symmetry_match_rate(current_path / "generated_crystals_cif", spg)
    if (current_path / "projected" / "generated_crystals_cif").exists():
        get_symmetry_match_rate(current_path / "projected" / "generated_crystals_cif", spg)        
```
